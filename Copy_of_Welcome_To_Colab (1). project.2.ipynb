```
import pandas as pd
import matplotlib.pyplot as plt

```

```
colors = ["#89CFF0", "#FF69B4", "#FFD700", "#7B68EE", "#FF4500",
          "#9370DB", "#32CD32", "#8A2BE2", "#FF6347", "#20B2AA",
          "#FF69B4", "#00CED1", "#FF7F50", "#7FFF00", "#DA70D6"]
```

```
df = pd.read_excel('/content/Project.2.(1)(1).xlsx')
print (df.head(10))
```

```
df.dropna(inplace=True)
```

```
df['ORDER DATE'] = pd.to_datetime(df['ORDER DATE'])
```

```

df['MONTH'] = df['ORDER DATE'].dt.month

df['QUANTITY ORDERED'] = pd.to_numeric(df['QUANTITY ORDERED'], errors='coerce')
df['PRICE EACH'] = pd.to_numeric(df['PRICE EACH'], errors='coerce')
df['SALES'] = df['QUANTITY ORDERED'] * df['PRICE EACH']
```

```
MONTHLY_SALES = df.groupby('MONTH')['SALES'].sum(10)

print(MONTHLY_SALES)
```

```
MONTHS = range(1, 13)
SALES_IN_USD = range(40,180)
plt.bar = df.groupby('MONTH')['SALES']
plt.xticks(MONTHS)
plt.xlabel('MONTHS')
plt.ylabel('SALES IN USD')
plt.title('MONTHLY SALES ANALYSIS')
plt.show()
```

```

df.columns = df.columns.str.strip()
```

```
PRODUCT_group = df.groupby('PRODUCT')['QUANTITY ORDERED'].sum()

# Get list of products
PRODUCT  = PRODUCT_group.index.tolist()

# Plot
plt.bar(PRODUCT, PRODUCT_group)
plt.xticks(rotation='vertical', fontsize=8)
plt.xlabel('Product')
plt.ylabel('QUANTITY ORDERED')
plt.title('Sales by Product')
plt.show()
```
